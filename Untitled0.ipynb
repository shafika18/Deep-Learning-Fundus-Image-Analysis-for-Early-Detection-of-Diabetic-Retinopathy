```python
from tensorflow.keras.layers import Dense, Flatten, Input
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img
from tensorflow.keras.applications.xception import Xception, preprocess_input

from glob import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_datagen = ImageDataGenerator(
    rescale = 1./255,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True
)

test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
import kagglehub
path = kagglehub.dataset_download("arbethi/diabetic-retinopathy-level-detection")
print(path)
```

```python
import os

print(os.listdir(path))
```

```python
base_path = os.path.join(
    path,
    "preprocessed dataset",
    "preprocessed dataset"
)

trainPath = os.path.join(base_path, "training")
testPath  = os.path.join(base_path, "testing")

print("Train path:", trainPath)
print("Test path :", testPath)
```

```python
print(os.listdir(trainPath))
```

```python
training_set = train_datagen.flow_from_directory(
    trainPath,
    target_size=(299, 299),
    batch_size=32,
    class_mode='categorical'
)

test_set = test_datagen.flow_from_directory(
    testPath,
    target_size=(299, 299),
    batch_size=32,
    class_mode='categorical'
)
```

```python
# Load Xception model WITHOUT top layer
xception = Xception(
    input_shape=(299, 299, 3),
    weights='imagenet',
    include_top=False
)
```

```python
# Freeze all layers
for layer in xception.layers:
    layer.trainable = False
```

```python
x = Flatten()(xception.output)

prediction = Dense(
    5,              # number of classes
    activation='softmax'
)(x)
```

```python
model = Model(
    inputs=xception.input,
    outputs=prediction
)
```

```python
model.summary()
```

```python
model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
```

```python
history = model.fit(
    training_set,
    validation_data=test_set,
    epochs=5,
    steps_per_epoch=len(training_set),
    validation_steps=len(test_set)
)
```
